### Vectors ###
arrays

```python
v=np.array([1,0,0])
u=np.array([0,-2,0])
```

```python
print(v,u)
```

Operacions amb vectors (suma, resta, producte escalar i vectorial)

```python
w_s= u+v
w_r=u-v
print(w_s, w_r)
```

```python
escalar = np.dot(u,v)
print(escalar)
```

```python
vectorial = np.cross(u,v)
print(vectorial)
```

### Diagonalitzar, valors propis i vectors propis ###
per diagonalitzar han de ser cuadrades!
numpy treballa amb num complexos, pel que pot diagonalitzar qualsevol.

```python
A=np.array([[2,2],[1,3]])
```

```python
valors_propis,vectors_propis= np.linalg.eig(A)
print("Val propis", valors_propis)
print("Vec propis", vectors_propis)
```

```python
Vp_1=vectors_propis[:,0]
Vp_2=vectors_propis[:,1]
print("Vp 1", Vp_1)
print("Vp 2", Vp_2)
```

```python
# Comprovacio que son vectors propis

z= np.dot(A,Vp_1)
print(z)

z_1= np.dot(A,Vp_2)
print(z_1)
```
