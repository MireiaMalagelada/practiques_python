#### Derivada ####

Podem definir 3 derivades, la forward, la central i la backward.

Jo fare servir la forward.

```python
# definicio de la funcio derivada.

def funcio_derivada(f,dx):
    valors_F=[]

    for i in range(0, len(f)-1):  # li haig de restar 1 a la llargada de la llista, perque no podra calcular la derivada de l'ultim valor
         F=(f[i+1]-f[i])/dx    # funció que calcula la derivada
         print("i= ",i," f[i]= ",f[i]," F= ",F)  # amb aixo puc comprobar que ho estigui fent be
         valors_F.append(F)
         i= i+1
    return(valors_F)
```

```python
# defineixo la funcio f_x y calculo els valors cada dx.

def funcio(x):
    fun_x= x**(3)   # aquesta es la nostre funcio
    return(fun_x)


dx=1
valors_f= []  # aqui guardo els valors de la funcio
valors_x=[]   # aqui guardo els valors de les x, aquest nomes servira per pintar la grafica

x_inicial=-10
x_final= 10

x=x_inicial
while x <= x_final:
    valors_x.append(x)
    f_x=funcio(x)
    print("x= ",x, " f_x= ",f_x)  # comprobacio de que esta fent be
    valors_f.append(f_x)
    x=x+dx

print("valors_f= ",valors_f)
print("valors_x= ",valors_x)
```

```python
# grafico f en funcio de x
grafic_f_x = plt.plot(valors_x, valors_f, 'ro')
plt.show(grafic_f_x)
```

```python
# derivada de la funcio
val_f_derivada = funcio_derivada(valors_f,dx)
print("derivada_f = ", val_f_derivada)
```

*Verlet Algorithm.* Let us consider that the function F(x) is the position of a particle as a function of time, **r_i(t)**, so the first derivative is the velocity **v_i(t)** and the second derivative is the acceleration **a_i(t)**. Using the results of Exercise 1 for the first and second derivative, show that:

vi(t) = ( ri(t + ∆t) − ri(t − ∆t) ) / 2∆t

ri(t + ∆t) = 2ri(t) − ri(t − ∆t) + ai(t)∆t^2


**Note:** This algorithm is more useful than the one from Feynman’s book (hopefully) seen at class that gives positions and velocities at different times. If we know the state of a particle at a given time, we can use these equations and Newton’s Second Law **F_i(t) = m_i · a_i(t)** to compute the trajectory of the particle over time in small time steps ∆t.

Both the position and velocity (and hence kinetic and potential energies) can be evaluated at the same instants of time.

Començarem definint les funcions com les tenim en el full:

```python
def posicio(x):
  f_posicio= funcio(x)

  return(f_posicio)

def velocitat(x,dx):
  f_velocitat= ( posicio(x+dx)-posicio(x-dx) )/(2*dx)

  return(f_velocitat)

def acceleracio(x, dx):
  f_acceleracio= ( posicio(x+dx)-2*posicio(x)+posicio(x-dx) )/ dx**2

  return(f_acceleracio)
```

#### Segona part ####

Passem a la segona part del exercici. (Diagonalitzar)

Comencem definint coses.

```python
# boundary conditions
val_f_ona[0]=0
val_f_ona[N-1]=0


# funcio a trossos que defineix els valors de la matriu F

def funcio_a_trossos(i,j):
  if j==i :
    F_ij= -2/nu -2*h**2 *potencial_1D(valors_x[i])

  if j==i-1 and i>1 :
    F_ij= 1/nu

  if j==i+1 and i<N :
    F_ij= 1/nu

  else:
    F_ij=0

  return(F_ij)

```

Ara muntem la matriu F

```python
# matriu F
F=[]

for i in range(0,N):  # i es el contador de la posicio del vector F i j es el dels subvectors
  F.append([])
  for j in range(0,N):
    F_ij= funcio_a_trossos(i,j)
    F[i].append(F_ij)
    print("i= ",i," j= ",j,"F_ij= ", F_ij)

F_array=np.array(F)
print("F= ",F)
print("\n ----------------------- \n")
print("F_array= ",F_array)
```

Ara busquem els valors propis i vectors propis de la matriu F_array

```python
# valors i vectors propis
valors_propis,vectors_propis= np.linalg.eig(F_array)
print("Val propis", valors_propis)
print("Vec propis", vectors_propis)
```
