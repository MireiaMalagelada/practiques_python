## Example parabolic motion

Example of use of numpy and Matplotlib to plot data for a problem with known analytical solution

```python
# Here we import the mathematical library and the plots library
import numpy as np
import matplotlib.pyplot as plt
```

*Physical parameters of the problem: acceleration of gravity (m/s2)*

```python
g=9.8 #m/s2
```

*Initial data: modulus of initial velocity (m/s) and angle (deg)*

```python
u=float(input('Enter initial velocity in m/s:')) #example 20 m/s
print(u)
```

```python
theta=float(input('Angle in degrees:')) # A good example is 45 degrees
```

```python
#convert angle in degrees to rad
theta = np.radians(theta)
print(theta, "rad")
```

**Solve the parabolic motion**

*Calculate time of flight*

```python
t_flight = 2*u*np.sin(theta)/g
```

```python
print(t_flight, "seconds")
```

*Divide the time in discrete intervals of size dt*

```python
dt=0.001 #seconds
```

```python
t = np.arange(0,t_flight,dt)
print(t)    
```

*Calculate the position x,y for each time in the interval*

```python
x=u*np.cos(theta)*t
```

```python
y=u*np.sin(theta)*t - (1.0/2.0)*g*t*t
```

```python
print(x)
```

```python
print(y)
```

Plot the results

```python
plt.plot(x, y)
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.title('Projectile motion')
plt.savefig("example_1.png")
plt.show()
```
